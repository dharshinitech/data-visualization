```python
def risk_level(score):  #risk category(high/medium/low)
    if score > 70:
        return "High"
    elif score >= 40:
        return "Medium"
    else:
        return "Low"

df["Risk_Level"] = df["AI_Risk_Score"].apply(risk_level)
```

```python
import matplotlib.pyplot as plt #pie chart for risk level

df["Risk_Level"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Risk Level Distribution")
plt.ylabel("")
plt.show()
```

```python
#salary vs ai risk scatter plot
plt.scatter(df["Salary"], df["AI_Risk_Score"])
plt.xlabel("Salary")
plt.ylabel("AI Risk Score")
plt.show()
```
